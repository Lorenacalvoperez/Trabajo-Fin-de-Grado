```python
import requests
import json

# URLs de la API
DATA_URL = "https://api.ourworldindata.org/v1/indicators/916408.data.json"
METADATA_URL = "https://api.ourworldindata.org/v1/indicators/916408.metadata.json"

# Obtener los datos
data_response = requests.get(DATA_URL).json()
metadata_response = requests.get(METADATA_URL).json()

# Guardar los datos en un archivo JSON para analizarlos
with open("data.json", "w") as f:
    json.dump(data_response, f, indent=4)

with open("metadata.json", "w") as f:
    json.dump(metadata_response, f, indent=4)

print("Datos guardados en data.json y metadata.json")
```

```python
import pandas as pd
import requests

# URLs de la API
DATA_URL = "https://api.ourworldindata.org/v1/indicators/916408.data.json"
METADATA_URL = "https://api.ourworldindata.org/v1/indicators/916408.metadata.json"

# Funcin para obtener y estructurar los datos
def fetch_data():
    # Obtener datos desde las APIs
    data_response = requests.get(DATA_URL).json()
    metadata_response = requests.get(METADATA_URL).json()

    # Extraer la lista de valores
    values = data_response["values"]

    # Extraer los aos desde metadata.json
    years = [year["id"] for year in metadata_response["dimensions"]["years"]["values"]]

    # Extraer los pases
    countries = {c["id"]: c["name"] for c in metadata_response["dimensions"]["entities"]["values"]}

    # Verificar si hay ms de un pas en los datos
    num_years = len(years)
    num_values = len(values)
    
    if num_values % num_years == 0:
        num_countries = num_values // num_years
        print(f"Se detectaron datos de {num_countries} pases")

        # Crear un DataFrame con mltiples pases
        data = []
        for i, (year, value) in enumerate(zip(years * num_countries, values)):
            country_id = list(countries.keys())[i // num_years]  # Asignar pas de forma cclica
            data.append({"Year": year, "Cases": value, "Country": countries[country_id]})

        df = pd.DataFrame(data)
    else:
        print("Los datos no coinciden perfectamente con los aos. Solo se usar un subconjunto.")
        df = pd.DataFrame({"Year": years[:num_values], "Cases": values[:num_values]})
        df["Country"] = "World"

    return df

# Ejecutar la funcin y mostrar resultados
df = fetch_data()
print(df.head())
```

## Variable de calidad de agua (nitrato)

```python
import requests
import json

# URLs de las APIS
DATA_URL = "https://api.ourworldindata.org/v1/indicators/820661.data.json"
METADATA_URL = "https://api.ourworldindata.org/v1/indicators/820661.metadata.json"

# Obtener los datos
data_json = requests.get(DATA_URL).json()  
metadata_json = requests.get(METADATA_URL).json()  

# Guardar los datos en un archivo JSON
with open("nitratos_data.json", "w") as f:
    json.dump(data_json, f, indent=4)

with open("nitratos_metadata.json", "w") as f:
    json.dump(metadata_json, f, indent=4)

print("Datos guardados en nitratos_data.json y nitratos_metadata.json")
```

### Procesamiento de los datos

```python
# Extraer aos y valores de los JSON
years = [year["id"] for year in metadata_json["dimensions"]["years"]["values"]]
values = data_json["values"]  # Lista de valores de concentracin de nitratos

# Verificar la longitud de los datos
print(f"Longitud de years: {len(years)}")
print(f"Longitud de values: {len(values)}")
```

```python
import pandas as pd

# Extraer valores de concentracin de nitratos
values = data_json["values"]

# Extraer aos desde metadata_json
years = [year["id"] for year in metadata_json["dimensions"]["years"]["values"]]

# Extraer pases desde metadata_json
countries = {c["id"]: c["name"] for c in metadata_json["dimensions"]["entities"]["values"]}

# Crear lista para almacenar los datos estructurados
structured_data = []

# Recorrer valores y asignarlos a pases y aos
value_index = 0

for country_id, country_name in countries.items():
    for year in years:
        if value_index < len(values):  
            structured_data.append({
                "Country": country_name,
                "Year": year,
                "Nitrate_Concentration_mgL": values[value_index]
            })
            value_index += 1
        else:
            structured_data.append({
                "Country": country_name,
                "Year": year,
                "Nitrate_Concentration_mgL": None
            })

# Convertir la lista en un DataFrame de pandas
df = pd.DataFrame(structured_data)

# Mostrar las primeras filas para verificar
print(df.head(20))
```

## Variable de exposicion al plomo

```python
import requests
import json

# URLs de la API
DATA_URL = "https://api.ourworldindata.org/v1/indicators/941463.data.json"
METADATA_URL = "https://api.ourworldindata.org/v1/indicators/941463.metadata.json"

# Obtener los datos
data_response = requests.get(DATA_URL).json()
metadata_response = requests.get(METADATA_URL).json()

# Guardar los datos en un archivo JSON para analizarlos
with open("plomo_data.json", "w") as f:
    json.dump(data_response, f, indent=4)

with open("plomo_metadata.json", "w") as f:
    json.dump(metadata_response, f, indent=4)

print("Datos guardados en data.json y metadata.json")
```

### Procesamiento de datos

**Verificar la estructura y el tamao de los datos**

```python
# Calcular la longitud de las estructuras clave
longitud_values = len(data_response["values"])
longitud_years = len(metadata_response["dimensions"]["years"]["values"])
longitud_entities = len(metadata_response["dimensions"]["entities"]["values"])

# Mostrar resultados
print(f"Longitud de 'values' en plomo_data.json: {longitud_values}")
print(f"Longitud de 'years' en plomo_metadata.json: {longitud_years}")
print(f"Longitud de 'entities' en plomo_metadata.json: {longitud_entities}")
```

**Identificar donde se encuentran las diferencias**

**ENTIDADES**

```python
# Extraer lista de entidades de la metadata
metadata_entities = {e["id"] for e in metadata_response["dimensions"]["entities"]["values"]}

# Extraer lista de entidades de los datos
data_entities = set(data_response["entities"])

# Encontrar diferencias
entities_missing_in_metadata = data_entities - metadata_entities
entities_missing_in_data = metadata_entities - data_entities

print(f"Nmero de entidades en los datos pero no en la metadata: {len(entities_missing_in_metadata)}")
print(f"Lista de entidades faltantes en metadata: {entities_missing_in_metadata}")

print(f"Nmero de entidades en la metadata pero no en los datos: {len(entities_missing_in_data)}")
print(f"Lista de entidades faltantes en los datos: {entities_missing_in_data}")
```

**AOS**

```python
# Extraer lista de aos de la metadata
metadata_years = {y["id"] for y in metadata_response["dimensions"]["years"]["values"]}

# Extraer lista de aos de los datos
data_years = set(data_response["years"])

# Encontrar diferencias
years_missing_in_metadata = data_years - metadata_years
years_missing_in_data = metadata_years - data_years

print(f"Nmero de entidades en los datos pero no en la metadata: {len(years_missing_in_metadata)}")
print(f"Lista de entidades faltantes en metadata: {years_missing_in_metadata}")

print(f"Nmero de entidades en la metadata pero no en los datos: {len(years_missing_in_data)}")
print(f"Lista de entidades faltantes en los datos: {years_missing_in_data}")
```
