## GLM

```python
df_entrenamiento = pd.read_csv("Tabla_filtrada.csv")

# Definir variables
variables_independientes = ['Contaminacion_aire', 'Muertes_agua', 'Exp_plomo', 'PeSticidas', 'Precipitaciones']
variable_dependiente = 'Parkinson'

# Diccionario para almacenar modelos por pas
modelos_por_pais_GLM = {}

# Entrenar modelos para cada pas
for pais in df_entrenamiento["Pas"].unique():
    print(f"Entrenando modelo para {pais}...")
    
    df_pais = df_entrenamiento[df_entrenamiento["Pas"] == pais].drop(columns=["Pas"])
    #display(df_pais)
    
    modelo, scaler, columnas_utilizadas = entrenar_modelo_glm(df_pais, sm.families.NegativeBinomial(), variables_independientes, variable_dependiente, scaler=True)
    # Guardar el modelo y el scaler en el diccionario
    modelos_por_pais_GLM[pais] = (modelo, scaler, columnas_utilizadas)
```

## Random Forest

```python
# Cargar los datos
df_entrenamiento = pd.read_csv("Tabla_filtrada.csv")

# Definir variables
variables_independientes = ['Contaminacion_aire', 'Muertes_agua', 'Exp_plomo', 'Pesticidas', 'Precipitaciones']
variable_dependiente = 'Parkinson'

# Diccionario para guardar los modelos por pas
modelos_por_pais_RF = {}

# Entrenar un modelo para cada pas
for pais in df_entrenamiento["Pas"].unique():
    print(f"\nEntrenando modelo para {pais}...")

    df_pais = df_entrenamiento[df_entrenamiento["Pas"] == pais].drop(columns=["Pas"])

    modelo, columnas_utilizadas = entrenar_modelo_rf(df_pais, variables_independientes, variable_dependiente)

    modelos_por_pais_RF[pais] = (modelo, columnas_utilizadas)
```
