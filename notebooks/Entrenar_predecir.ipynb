```python
import os
import pandas as pd

def cargar_csv(nombre_archivo):
    """
    Carga un archivo CSV desde la ruta 'Resultados/Archivos_csv/Datos' en un DataFrame de pandas.

    Parámetros
    ----------
    nombre_archivo : str
        Nombre del archivo CSV a cargar.

    Retorna
    -------
    pd.DataFrame
        DataFrame con los datos cargados del archivo CSV.
    """
    # Construir la ruta completa
    ruta_archivo = os.path.join("..","Resultados", "Archivos_csv", "Datos", nombre_archivo)
    
    # Leer el CSV
    df = pd.read_csv(ruta_archivo, encoding="utf-8")
    print(f"Archivo '{ruta_archivo}' cargado correctamente.")
    
    return df
```

```python
import sys
import os

# Agrega la carpeta 'src' al path de búsqueda
ruta_src = os.path.abspath(os.path.join(os.getcwd(), '..', 'src'))

if ruta_src not in sys.path:
    sys.path.insert(0, ruta_src)

from Entrenamiento_Modelos import *
from Prediccion_modelos import *

```

## Entrenamiennto de Modelos

## GLM

```python
df_entrenamiento = cargar_csv("Tabla_filtrada.csv")

# Definir variables
variables_independientes = ['Contaminacion_aire', 'Muertes_agua', 'Exp_plomo', 'Pesticidas','Precipitaciones']
variable_dependiente = 'Parkinson'

# Diccionario para almacenar modelos por país
modelos_por_pais_GLM = {}

# Entrenar modelos para cada país
for pais in df_entrenamiento["País"].unique():
    print(f"Entrenando modelo para {pais}...")
    
    df_pais = df_entrenamiento[df_entrenamiento["País"] == pais].drop(columns=["País"])
    #display(df_pais)
    
    modelo, scaler, columnas_utilizadas,formula = entrenar_modelo_glm(df_pais, sm.families.NegativeBinomial(), variables_independientes, variable_dependiente, scaler=True)
    # Guardar el modelo y el scaler en el diccionario
    modelos_por_pais_GLM[pais] = (modelo, scaler, columnas_utilizadas,formula)
```

## Random Forest

```python
# Cargar los datos
df_entrenamiento = cargar_csv("Tabla_filtrada.csv")

# Definir variables
variables_independientes = ['Contaminacion_aire', 'Muertes_agua', 'Exp_plomo', 'Pesticidas','Precipitaciones']
variable_dependiente = 'Parkinson'

# Diccionario para guardar los modelos por país
modelos_por_pais_RF = {}

# Entrenar un modelo para cada país
for pais in df_entrenamiento["País"].unique():
    print(f"\nEntrenando modelo para {pais}...")

    # Filtrar datos por país
    df_pais = df_entrenamiento[df_entrenamiento["País"] == pais].drop(columns=["País"])

    # Llamar a la función para entrenar y evaluar el modelo
    modelo,variables_independientes = entrenar_modelo_rf(df_pais, variables_independientes, variable_dependiente)

    # Guardar el modelo para cada país
    modelos_por_pais_RF[pais] = (modelo,variables_independientes)
```

```python
# Renombramos y nos quedamos solo con Pais + predicción
df1 = df_predicciones_GLM[['País', 'Parkinson_Predicho']].rename(columns={'Parkinson_Predicho': 'GLM'})
df2 = df_predicciones_RF[['País', 'Parkinson_Predicho']].rename(columns={'Parkinson_Predicho': 'RF'})
df3 = df_predicciones_XG[['País', 'Parkinson_Predicho']].rename(columns={'Parkinson_Predicho': 'XG'})
df4 = df_predicciones_SVR[['País', 'Parkinson_Predicho']].rename(columns={'Parkinson_Predicho': 'SVR'})
df5 = df_predicciones_KNN[['País', 'Parkinson_Predicho']].rename(columns={'Parkinson_Predicho': 'KNN'})
df6 = df_predicciones_MLP[['País', 'Parkinson_Predicho']].rename(columns={'Parkinson_Predicho': 'MLP'})

# Hacemos el merge progresivo sin columnas que se repitan
df_predicciones_modelos = df1.merge(df2, on='País') \
               .merge(df3, on='País') \
               .merge(df4, on='País') \
               .merge(df5, on='País') \
               .merge(df6, on='País')


# Paso 3: Calcular media, desviación estándar y CV
df_predicciones_modelos['Parkinson_Predicho_Promedio'] = df_predicciones_modelos[['GLM', 'RF', 'XG', 'SVR', 'KNN', 'MLP']].mean(axis=1)
df_predicciones_modelos['Desviacion'] = df_predicciones_modelos[['GLM', 'RF', 'XG', 'SVR', 'KNN', 'MLP']].std(axis=1)


# 🔽 AQUÍ insertas el redondeo
df_predicciones_modelos[['GLM', 'RF', 'XG', 'SVR', 'KNN', 'MLP',
           'Parkinson_Predicho_Promedio', 'Desviacion']] = \
df_predicciones_modelos[['GLM', 'RF', 'XG', 'SVR', 'KNN', 'MLP',
           'Parkinson_Predicho_Promedio', 'Desviacion']].round(2)

# Paso 4: Guardar los tres archivos
# Guardar archivos usando la ruta definida previamente
df_predicciones_modelos[['País', 'GLM', 'RF', 'XG', 'SVR', 'KNN', 'MLP', 'Parkinson_Predicho_Promedio']].to_csv(
    f"{carpeta_csv}/predicciones_modelos_promedio.csv", index=False)

df_predicciones_modelos[['País', 'GLM', 'RF', 'XG', 'SVR', 'KNN', 'MLP', 'Desviacion']].to_csv(
    f"{carpeta_csv}/predicciones_modelos_desviacion.csv", index=False)


# Resultado final
print(df_predicciones_modelos[['País', 'Parkinson_Predicho_Promedio','Desviacion']])
```
